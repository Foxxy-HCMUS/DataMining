```python
import numpy as np
import pandas as pd
```

```python
hawks_df = pd.read_csv('./hawks.csv')
hawks_df.head(10)
```

```python
hawks_df.shape
```

```python
def check(df):
    for item in df.columns:
        print(df[item].dtype)
        print('<' + item + '>', ((df[item].isna()) | (df[item] == ' ')).value_counts(normalize=True)*100)
        print()

# kiểm tra về kiểu dữ liệu và các dữ liệu bị thiếu trong df 
check(hawks_df)
```

- Điền đầy dữ liệu số bằng mean và điền đầy dữ liệu định danh bằng mode  
- Đối với những thuộc tính bị thiếu trên 50% thì xoá luôn thuộc tính đi  
- Do đó các thuộc tính bị xoá sẽ là
    - ReleaseTime: thiếu 92.8%
    - Sex: thiếu 63.4%
    - Tarsus: thiếu 91.7%
    - WingPitFat: thiếu 91.5%
- Các thuộc tính sẽ được điền đầy bằng phương pháp mean
    - Wing 
    - Weight
    - Culmen 
    - Hallux 
    - StandardTail
    - KeelFat 
    - Crop
- Thuộc tính sẽ được điền đầy bằng phương pháp mode: CaptureTime 
- Quan sát dữ liệu cột BandNumber, ta thấy thuộc tính này hoàn toàn có thể là thuộc tính index vì tất cả các dòng dữ liệu đều khác nhau. Do đó, ta xoá luôn cột này

```python
# xoá cột
hawks_df.drop(columns=['ReleaseTime', 'Sex', 'Tarsus', 'WingPitFat', 'BandNumber'], inplace=True)
hawks_df.head()
```

Tại các cột dữ liệu định danh, có 2 cột 

```python
# điền đầy các cột numeric bằng mean 
num_col = ['Wing' ,'Weight','Culmen' ,'Hallux' ,'StandardTail','KeelFat' ,'Crop']

for item in num_col:
    hawks_df[item] = hawks_df[item].fillna(hawks_df[item].mean())

hawks_df.head()
```

Đối với các cột norminal bị thiếu dữ liệu, cột `CaptureTime`, `BandNumber`


```python
mode = hawks_df['CaptureTime'].mode()
hawks_df['CaptureTime'] = hawks_df['CaptureTime'].replace([np.nan, ' '], mode)
check(hawks_df)
```

```python
# lưu file
hawks_df.to_csv('afterPreprocess.csv', header=True, index=False)
```
